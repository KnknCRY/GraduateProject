```python
import wave
import contextlib
import math
import speech_recognition
from os import listdir
from os.path import isfile, isdir, join
from pydub import AudioSegment
import jieba
import re
import time
import multiprocessing.dummy as mp
import json
import datetime
import random

# 指定要列出所有檔案的目錄
mypath = "speech_voice_sliced" #所有音檔都在這
# 取得所有檔案與子目錄名稱
files = listdir(mypath)

info_dic = {} #id uid time lenth
barchart_dic = {"data":[], "filename":[[],[],[],[],[],[],[],[]]}
make_json = 0

a = 0
b = 0
c = 0
d = 0
e = 0
f = 0
g = 0
h = 0

for fname in files:
    make_json += 1
    if make_json%2 != 0:#音檔成對，但我只要做其中一個
        t = fname.split("-")[1]
        t = int(int(t)/1000)
        s = datetime.datetime.fromtimestamp(t)

        if s.hour < 3 and s.hour >= 0:
            a += 1
            barchart_dic["filename"][0].append("./data/info/"+fname[:-5]+"_info.json")
        elif s.hour < 6 and s.hour >= 3:
            b += 1
            barchart_dic["filename"][1].append("./data/info/"+fname[:-6]+"_info.json")
        elif s.hour < 9 and s.hour >= 6:
            c += 1
            barchart_dic["filename"][2].append("./data/info/"+fname[:-6]+"_info.json")
        elif s.hour < 12 and s.hour >= 9:
            d += 1
            barchart_dic["filename"][3].append("./data/info/"+fname[:-6]+"_info.json")
        elif s.hour < 15 and s.hour >= 12:
            e += 1
            barchart_dic["filename"][4].append("./data/info/"+fname[:-6]+"_info.json")
        elif s.hour < 18 and s.hour >= 15:
            f += 1
            barchart_dic["filename"][5].append("./data/info/"+fname[:-6]+"_info.json")
        elif s.hour < 21 and s.hour >= 18:
            g += 1
            barchart_dic["filename"][6].append("./data/info/"+fname[:-6]+"_info.json")
        else:
            h += 1
            barchart_dic["filename"][7].append("./data/info/"+fname[:-6]+"_info.json")
            
        
        #開始做info_dic
        sound = AudioSegment.from_file(mypath + "\\" + fname)
        info_dic["id"] = fname[:-6]
        
        if (fname.split("-")[0]=="Room_0601553551230" or fname.split("-")[0]=="Room_1441294301270" or 
        fname.split("-")[0]=="Room_2283455301270" or fname.split("-")[0]=="Room_4553444981220" or 
        fname.split("-")[0]=="Room_5732276261220" or fname.split("-")[0]=="Room_8567569881220" or 
        fname.split("-")[0]=="Room_9440395301270" or fname.split("-")[0]=="Room_86548202112110" or 
        fname.split("-")[0]=="Room_87395691112250") :
            info_dic["uid"] = "17801112191240"
            print("17801112191240")
        elif (fname.split("-")[0]=="Room_03154134902250" or fname.split("-")[0]=="Room_26151354902300" or 
             fname.split("-")[0]=="Room_27309534902300" or fname.split("-")[0]=="Room_37820803012300" or 
             fname.split("-")[0]=="Room_39611213012250" or fname.split("-")[0]=="Room_65020644902300" or 
             fname.split("-")[0]=="Room_66895713012250" or fname.split("-")[0]=="Room_80683267491190"):
            info_dic["uid"] = "73140858551180"
            print("73140858551180")
        else:
            info_dic["uid"] = "97409473651160"
            print("97409473651160")
            
        if s.minute < 10:
            info_dic["time"] = str(s.hour)+":0"+str(s.minute)
        else:
            info_dic["time"] = str(s.hour)+":"+str(s.minute)
        info_dic["lenth"] = str(int(len(sound)/1000))
        info_dic["link"] = ["./data/radar/"+fname[:-6]+"_radar.json", "./data/wordcloud/"+fname[:-6]+"_wordcloud.json",
                           "./data/eventdrops/"+fname[:-6]+".json"]
        info_dic["satisfied"] = random.randint(1,5)
        info_dic["keywords"] = []
        timeword_path = "speech_voice_timeword_json"
        with open(timeword_path+"\\"+fname[:-6]+"_timeword.json", 'r',encoding="utf-8") as reader:
            jf = json.loads(reader.read())
        t = 0
        for x in jf.keys():
            if t > 2:
                break
            info_dic["keywords"].append(x)
            t += 1
        with open('speech_voice_info_json'+'\\'+fname[:-6]+'_info.json', "w",encoding='utf-8') as F:
            json.dump(info_dic,F,ensure_ascii=False,sort_keys=True, indent=4)
        
barchart_dic["data"] = [a,b,c,d,e,f,g,h]
with open('speech_voice_barchart_json'+'\\'+'barchart.json', "w",encoding='utf-8') as F:
    json.dump(barchart_dic,F,ensure_ascii=False,sort_keys=True, indent=4)
```
